```python
def play_again():
    return tic_tac_toe()

def tic_tac_toe(start='Y'):
    # Initialize variables
    player_1 = ''
    x_mark = ["d", " ", " ", " ", " ", " ", " ", " ", " ", " "]
    active_player = 1
    won = False
    full = False
    options = [1,2,3,4,5,6,7,8,9]

    # Introduction and start question
    print('Welcome to Tic Tac Toe!\n')
    player_1 = input('Player 1: Do you want to be X or O?  ')

    # Validate input, ask again if incorrect
    while player_1.upper() != 'X' and player_1.upper() != 'O':
        player_1 = input('That is not an option. Player 1: Do you want to be X or O?  ')

    # After a valid mark for player1, then set mark for player 2
    if player_1 == 'X':
        player_2 = 'O'
    else:
        player_2 = 'X'

    # Input first round
    while won == False and full == False:
        while True:
            round = input(f'Player {active_player}: Please choose an open position (between 1 and 9)  ')
            try:
                round = int(round)

                if round not in options or x_mark[round] != ' ':
                    print(f'Invalid input or position already taken. Please choose an open position (between 1 and 9).')
                else:
                    break

            except ValueError:
                print("Invalid input. Please enter a numeric value.")

        if active_player == 1:
            x_mark[int(round)] = player_1
        else:
            x_mark[int(round)] = player_2

        print('OK, great move, this is what the board looks like now: \n')
        print(f'  {x_mark[7]}  |  {x_mark[8]}  |  {x_mark[9]}  ')
        print('-----------------')
        print(f'  {x_mark[4]}  |  {x_mark[5]}  |  {x_mark[6]}  ')
        print('-----------------')
        print(f'  {x_mark[1]}  |  {x_mark[2]}  |  {x_mark[3]}  \n')

        if (x_mark[1]!= ' ' and x_mark[1] == x_mark[2] and x_mark[2] == x_mark[3]) or (x_mark[4]!= ' ' and x_mark[4] == x_mark[5] and x_mark[5] == x_mark[6]) or (x_mark[7]!= ' ' and x_mark[7] == x_mark[8] and x_mark[8] == x_mark[9]) or (x_mark[1]!= ' ' and x_mark[1] == x_mark[4] and x_mark[4] == x_mark[7]) or (x_mark[2]!= ' ' and x_mark[2] == x_mark[5] and x_mark[5] == x_mark[8]) or (x_mark[3]!= ' ' and x_mark[3] == x_mark[6] and x_mark[6] == x_mark[9]) or (x_mark[7]!= ' ' and x_mark[7] == x_mark[5] and x_mark[5] == x_mark[3]) or (x_mark[1]!= ' ' and x_mark[1] == x_mark[5] and x_mark[5] == x_mark[9]):
            won == True
            print(f'Game is won by player {active_player}\n')
            replay = input("Do you want to play again? (Y/N) ")
            while replay.upper() != 'Y' and replay.upper() != 'N':
                replay = input('That is not an option. Do you want to play again? (Y/N)  ')
            if replay.upper() == 'Y':
                print(f'Replay is {replay}\n')
                play_again()
            
            break
        if ' ' not in x_mark:
            full = True
            print(f'End of the game. No one won....\n')
            replay = input("Do you want to play again? (Y/N) ")
            while replay.upper() != 'Y' and replay.upper() != 'N':
                replay = input('That is not an option. Do you want to play again? (Y/N)  ')
            if replay.upper() == 'Y':
                print(f'Replay is {replay}\n')
                play_again()
            
            break
        # Now it's time for the other player
        if active_player == 1:
            active_player = 2
        else:
            active_player = 1
        
```
